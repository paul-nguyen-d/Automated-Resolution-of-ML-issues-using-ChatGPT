```python
import base64
import requests
from bs4 import BeautifulSoup

ACCESS_TOKEN = "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"

def get_file_content(owner,repo,file_path):
    # set the API endpoint for getting the file contents
    file_url = f"https://api.github.com/repos/{owner}/{repo}/contents/{file_path}" 
    headers = {"Authorization": f"Bearer {ACCESS_TOKEN}"}

    # make a request to the GitHub API to get the file contents
    response = requests.get(file_url, headers=headers)
    # get the response as JSON
    file_info = response.json()
    # decode the file contents from base64 encoding
    file_content = base64.b64decode(file_info['content']).decode('utf-8')

    return file_content
```

```python
count = 0

for i, name in enumerate(df["Project Name"]):
    files_no = 0
    if df["GPT_ans"].iloc[i] == "" and len(eval(df["Files changed"].iloc[i]))<14 and len(eval(df["Files changed"].iloc[i]))>1 and count < 10:
        owner = name.split("/")[0]
        repo = name.split("/")[1]
        files = eval(df["Files changed"].iloc[i])

        responses = ""
        for file_path in files:
            try:
                content = get_file_content(owner, repo, file_path)
                prompts = [{"role": "system", "content": f'''This is a github issue content: {df["Body"].iloc[i]}.
                                            The files that cause the issue are {df["Files changed"].iloc[i]}.
                                            You will be given the code of each file, please provide the fix to each of them to solve the issue.
                                            Return the solution in diff format to show the difference between the file before and after the fix.'''},
                            {"role": "user", "content": f"{file_path}:\n{content}"}]
                # print(prompts)
                response = openai.ChatCompletion.create( 
                    model="gpt-4",
                    messages=prompts)
                reply = response["choices"][0]["message"]["content"]
                # prompts.append({"role": "assistant", "content": reply})
                responses += f"\n{file_path}:{reply}"

                print("Done "+file_path)
                files_no += 1
            except openai.error.InvalidRequestError:
                print(f"InvalidRequestError while processing {file_path}. Skip to next file.")
                continue
            except:
                print(f"Error while processing {file_path}. Skip to next file.")
                continue

        df["Fixed files"].iloc[i] = files_no
        df["GPT_ans"].iloc[i] = responses
        print("Done item "+str(i))
        count += 1
```

```python
import matplotlib.pyplot as plt

# Data
completeness = 3.9
clarity = 4.4

# Creating the horizontal bar chart
fig, ax = plt.subplots()

# Plotting the bars
ax.barh(['Completeness', 'Clarity'], [completeness, clarity])

# Adding data labels
for i, v in enumerate([completeness, clarity]):
    ax.text(v + 0.1, i, str(v), color='black', va='center')

# Formatting the chart
ax.set_xlabel("Rating")
ax.set_ylabel("Evaluation Criteria")
ax.set_title("Evaluation Ratings")

# Displaying the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
completeness = 3.9
clarity = 4.4
max_score = 5

# Creating the completeness pie chart
completeness_label = f'Completeness: {completeness}/{max_score}'
incompleteness = max_score - completeness

completeness_data = [completeness, incompleteness]
completeness_labels = [completeness_label, '']

fig, ax = plt.subplots()
ax.pie(completeness_data, labels=completeness_labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
plt.title("Completeness Rating")
plt.show()

# Creating the clarity pie chart
clarity_label = f'Clarity: {clarity}/{max_score}'
lack_of_clarity = max_score - clarity

clarity_data = [clarity, lack_of_clarity]
clarity_labels = [clarity_label, '']

fig, ax = plt.subplots()
ax.pie(clarity_data, labels=clarity_labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
plt.title("Clarity Rating")
plt.show()
```
